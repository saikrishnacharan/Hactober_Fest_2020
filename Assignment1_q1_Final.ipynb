```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## **Part 1 a**

```python
fg1=cv.imread('../images/fg.jpg',1)
bg1=cv.imread('../images/bg.jpg',1)

fg1=cv.cvtColor(fg1,cv.COLOR_BGR2RGB)
bg1=cv.cvtColor(bg1,cv.COLOR_BGR2RGB)
fg=np.copy(fg1)
bg=np.copy(bg1)

f, axarr = plt.subplots(1,2,figsize=(15,10))
axarr[0].imshow(fg1)
axarr[1].imshow(bg1)
axarr[0].set_title("Original Fore Ground")
axarr[1].set_title("Original Back Ground")
```

```python
mask = cv.inRange(fg1, np.array([0,150,0]) ,np.array([180,255,150]))
# mask = cv.inRange(fg1, np.array([0,200,0]) ,np.array([150,255,100]))

fg[mask!=0]=[0,0,0]

bg[mask == 0] = [0, 0, 0]

final_image=bg+fg

f, axarr = plt.subplots(2,2,figsize=(15,10))
axarr[0,0].imshow(mask,cmap="gray")
axarr[0,1].imshow(fg)
axarr[1,0].imshow(bg)
axarr[1,1].imshow(final_image)
axarr[0,0].set_title("Mask")
axarr[0,1].set_title("Changed ForeGround")
axarr[1,0].set_title("Changed BackGround")
axarr[1,1].set_title("Final Image")
```

 Using CvinRange Function ,if it does fall in this range, the mask will be allowed to be displayed (**WHITE**) and if not it will block it out and turn the pixel (**BLACK**).
 
 Out motto is to make green **- ->** White and Figure **-->** Black of the Original Fore Ground Image
